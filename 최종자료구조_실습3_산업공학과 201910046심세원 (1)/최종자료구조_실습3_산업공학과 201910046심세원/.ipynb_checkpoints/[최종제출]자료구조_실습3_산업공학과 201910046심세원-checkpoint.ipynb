# 문제 1

```python
class ArrayList:
    def __init__(self):
        self.items=[]
    def add(self,elem):
        self.items.append(elem)
        return self.items
        
    def remove(self,elem):
        if elem in self.items:
            self.items.remove(elem)
            return print(elem, "removed")
        else:
            return print("No such element")
            
    def insert(self,pos, elem):
        self.items.insert(pos, elem)
        
    def delete(self):
        new_list=[]
        for i in self.items:
            if i not in new_list:
                new_list.append(i)
      
        self.items=new_list
        return self.items
    
    
    
    def isEmpty(self):
        return print(len(self.items)==0)
    
    def getEntry(self,pos):
        return print(self.items[pos])
    
    def size(self):
        return len(self.items)
    
    def clear(self):
        self.items=[]
        
    def find(self,elem):
        return self.items.index(elem)
    
    def replace(self,pos,elem):
        self.items[pos]=elem
        
    def sort(self):
        self.items.sort()
        
    def merge(self,lst):
        self.items.extend(lst)
        
    def search(self, elem):
        if elem in self.items:
            return print(elem, "found")
        else:
            return print("No such element")
            
    def printing(self,msg="ArrayList"):
        return print(msg, self.size(), self.items)
```

```python
def myEditor():
    print("Enter a command: i(insert), d(elete), e(mpty), g(etEntry), c(lear), a(dd),dup, remove, search, f(ind), r(eplace), s(ort),p(rint):")
    e=ArrayList()
    
    while True:
        c=[]
        c=input().split()
        
        if c[0]=='a':
            e.add(c[1])
            
        if c[0]=='p':
            e.printing()
            
        if c[0]=='i':
            e.insert(int(c[1]),int(c[2]))
            
        if c[0]=='d':
            e.delete(c[1]) 
            
        if c[0]=='e':
            e.isEmpty()
            
        if c[0]=='g':
            e.getEntry(int(c[1]))
            
        if c[0]=='c':
            e.clear()
            
        if c[0]=='dup':
            e.delete()
            
        if c[0]=='remove':
            e.remove(c[1])  
            
        if c[0]=='search':
            e.search(c[1])
            
        if c[0]=='f':
            e.find()
            
        if c[0]=='r':
            e.replace(int(c[1]),c[2])
            
        if c[0]=='s':
            e.sort()
            
        if c[0]=='m':
            e.merge(c[1:])
            
        if c[0]=='q':
            break
myEditor()    
```

# 문제 3

#### isSubset(=isSubstring())의 경우
집합A에 대한 부분집합(subset)리스트를 별도로 만들고
그 리스트에 대해 집합B와 비교하여
집합A에 대한 부분집합을 모두 담은 리스트안에 집합B(비교하려는 집합)가 있으면 True, 없으면 False를 출력하는 것을 구현하려고 했습니다.

#### isProperSubstring()의 경우
부분집합이지만 equal이 아닌 경우에 True, 그 외를 False로 구현하려고 했습니다.

```python
class Set:
    def __init__(self):
        self.items=[]
    
    def __eq__(self, setB):
        return self.items==setB
    
    def insert(self, elem):
        if elem not in self.items:
            self.items.append(elem)
        return self.items
        
    def delete(self, elem):
        if elem in self.items:
            self.items.remove(elem)
        return self.items
    

    def union(self, B):
        C=Set()
        C.items=list(self.items)
        for elem in B.items:
            if elem not in self.items:
                C.items.append(elem)
        return C.items
                
    def intersect(self, B):
        C=Set()
        for elem in B.items:
            if elem in self.items:
                C.items.append(elem)
        return C.items
        
    def difference(self, B):
        C=Set()
        for elem in self.items:
            if elem not in B.items:
                C.items.append(elem)
        return C.items
    
    def isSubset(self, B):            
    #######부분집합만드는 코드############
        arr = self.items
        C = len(arr)
        subset = [[]] 
        for num in range(C):
            size  = len(subset)
            for y in range(size):
                subset.append(subset[y] + [num])
    #######################################
        for elem in self.items:
            if elem in subset:
                return print("True")
            else:
                return print("False")            
    
#     def isProperSubset(self, setB):
#         return 
        
    def size(self):    
        return len(self.items)
        
#     def printing(self, msg): 
#         return print()
    
#     def isSubstring(self):
        
    
#     def isProperSubstring(self):
        
        
def test(A, B):

    print("SetA:",A.items)
    print("SetB:",B.items)
    if A==B:
        print("A equal B: True")
    else:
        print("A equal B: False")
        
    print("A subset B:", A.isSubset(B))
    #print("A proper subset B:", setA.isProperSubset(setB))
    print("A union B:", A.union(B))
    print("A intersect B:", A.intersect(B))
    print("A difference B:", A.difference(B))
    print()
```
