```python
def main():
    t=BSTree()
    print("Enter a command: i(nsert), s(earch), d(elete), p(rint), or q(uit)")
    while True:
        print(">", end='')
        line=input().split()
        command=line[0]
        
        if command=='i':
            item=int(line[1])
            if t.insert(item):
                print(item, " is inserted")
            else:
                print("Not inserted: same data")

        elif command=='s':
            item=int(line[1])
            if t.search(item):
                print(item, " is found")
            else:
                print("No such item in the tree")
        
        elif command=='d':
            item=int(line[1])
            if t.delete(item):
                print(item, " is deleted")
            else:
                print("No such item in the tree")

        elif command=='p':
            t.print()
        elif command=='q':
            break
```
