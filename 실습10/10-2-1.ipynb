```python
class BSTNode:
    def __init__(self, item, left=None, right=None):
        self.item=item
        self.left=left
        self.right=right
```

```python
class BSTree:
    def __init__(self):
        self.root=None #self.root = BSTree의 첫 번째 BST노드가 들어갑니다.
        self.parent=None
        

    def insert(self, data):
        
        #우선 첫 번째 insert인지 탐색.
        if self.root is None:
          #print("첫 번째 노드 삽입")
          self.root = BSTNode(data)

          #print("삽입 아이템 확인 : "  ,self.root.item)
          return True; #node가 제대로 삽입되었으면 True 리턴
        else:
          #print("두 번째 이상 노드 삽입")
          curr = self.root #현재 탐색할 curr Node 를 self.root(첫 번째 Node)로
          
          while curr is not None: #Node를 탐색하는 while
            #print(curr.item) #현재 탐색중인 노드의 값
            if curr.item is data: #만약 현재 탐색중인 Node의 아이템값이 data의 값과 일치한다면 false 리턴하여 중복 삽입을 막음.
              #print("중복 삽입된 데이터입니다")
              return False;
           #데이터 탐색
            if data < curr.item: #data가 작을땐 curr이 curr.left로
                if curr.left is None:
                  curr.left = BSTNode(data)
                  return True;
                #parent = curr
                curr = curr.left
            else : #data가 클 땐 curr이 curr.right로 이동
                if curr.right is None:
                  curr.right = BSTNode(data)
                  return True;
                #parent = curr
                curr = curr.right

          #while문을 빠져나오면 현재 curr == None
          
          #print(parent)#curr == None임을 확인




    def print(self):   #RVL 순회
        print()
        self.print1(self.root,0)


    def print1(self,n,count):
            
        if n is not None:
            count+=1
            self.print1(n.right,count)
            
           
            if n.right is not None and n.left is None:
                print(end='\t'*count )
                print(n.item,",R")
                #count+=1

            elif n.right is None and n.left is not None:
                print(end='\t'*count )
                print(n.item,",L")
                #count+=1
                
            elif n.right is not None and n.left is not None:
                print(end='\t'*count )
                print(n.item,",L,R")
                #count+=1
                
            else:
                #print(end='\t')
                print(end='\t'*count)
                print(n.item)
                #count+=1
                

            #print(end='\t')
        
            
            self.print1(n.left,count)
            count+=1     


    def search(self, data):
        curr=self.root
        while curr is not None:
            if data<curr.item:
                curr=curr.left
            elif data>curr.item:
                curr=curr.right
            else: return True
        return False
        

    def delete(self, data):
        #parent=None
        curr=self.root
        while curr is not None:
            if data<curr.item:
                self.parent=curr
                curr=curr.left
            elif data>curr.item:
                self.parent=curr
                curr=curr.right
            else:
                break
        if curr is None:
            return False

        if curr.left is None:
            if self.parent is None:
                self.root=curr.right
            else:
                if data<self.parent.item:
                    self.parent.left=curr.right
                else:
                    self.parent.right=curr.right

        elif curr.right is None:
            if self.parent is None:
                self.root=curr.left
            else:
                if data<self.parent.item:
                    self.parent.left=curr.left
                else:
                    self.parent.right=curr.left
                    
        else:
            parentMaxNode=curr
            maxNode=curr.left
            while maxNode.right is not None:
                parentMaxNode=maxNode
                maxNode=maxNode.right
            curr.item=maxNode.right
        return True
```

```python
def main():
    t=BSTree()
    print("Enter a command: i(nsert), s(earch), d(elete), p(rint), or q(uit)")
    while True:
        print(">", end='')
        line=input().split()
        command=line[0]
        
        if command=='i':
            item=int(line[1])
            if t.insert(item):
                print(item, " is inserted")
            else:
                print("Not inserted: same data")

        elif command=='s':
            item=int(line[1])
            if t.search(item):
                print(item, " is found")
            else:
                print("No such item in the tree")
        
        elif command=='d':
            item=int(line[1])
            if t.delete(item):
                print(item, " is deleted")
            else:
                print("No such item in the tree")

        elif command=='p':
            t.print()
        elif command=='q':
            break
```

```python
main()
```
