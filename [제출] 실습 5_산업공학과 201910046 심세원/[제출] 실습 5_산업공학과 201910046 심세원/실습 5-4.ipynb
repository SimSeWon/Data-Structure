```python
class deque():
    def __init__(self):    
        self.items=[]
        self.front=0                    #인덱스
        self.rear=-1                    #인덱스
        self.count=0
        
    def enqueue(self, item):
        #self.rear=self.rear+1
        #self.items[self.rear]=item
        self.items.append(item)
        self.count+=1

    def __ge__(self):
        return self.items[0]>=self.items
        
            
    def deleteFront(self):
        if not self.isEmpty():
            #item=self.items[self.front]
            #self.front=self.front+1
            self.count-=1
            #self.items.pop(0)
            return self.items.pop(0)
        
    def deleteRear(self):
        if not self.isEmpty():
            #item=self.items[self.rear]
            #self.items.pop()
            #self.rear=self.rear-1
            self.count-=1
            return self.items.pop()
        
    def isEmpty(self):
        return self.count==0

    

        
```

```python
def main():
    inFile=open('input1.txt')
    words=[]
    num=0
    while True:
        line=inFile.readline().splitlines()
        if line=='':
            break
        
        for i in line:
            #print(i)
            words.append(i)
            spelling=deque()
            for j in i:
                spelling.enqueue(j)
            #print(spelling.items)
            #print(words[num])

                #print(spelling.deleteFront(), spelling.deleteRear())


        
        while True:
            
            frontSpelling = spelling.deleteFront()
            rearSpelling = spelling.deleteRear()
            if type(frontSpelling)==str and type(rearSpelling)==str:
                frontSpelling=frontSpelling.lower()
                rearSpelling=rearSpelling.lower()

            
                
            if frontSpelling != rearSpelling:
                print(words[num])
                print('==> not palindrome')
                num+=1
                break

            if len(spelling.items)<=1:
                print(words[num])
                print('==> palindrome')
                num+=1
                break
        if num==10:
                break

    inFile.close()            
            
              
    
     
```
