```python
class BTNode:
    def __init__(self, item, left=None, right=None):
        self.item=item
        self.left=left
        self.right=right

class BTree:
    def __init__(self):
        self.root=None

    def bulidTree(self):
        g=BTNode(12)
        d=BTNode(15,g)
        b=BTNode(10, None, d)
        e=BTNode(25)
        f=BTNode(55)
        c=BTNode(30,e,f)
        self.root=BTNode(20,b,c)



    
    def inorder(self):
        self.inorder1(self.root)
        print()

    def inorder1(self, n):
        if n is not None:
            self.inorder1(n.left)
            print(n.item, end=' ')
            self.inorder1(n.right)

        
    def preorder(self):        #전위 순회
        self.preorder1(self.root)
        print()

    def preorder1(self,n):        
        if n is not None:
            print(n.item, end=' ')
            self.preorder1(n.left)
            self.preorder1(n.right)


    def postorder(self):          #후위 순회
        self.postorder1(self.root)
        print()

    def postorder1(self,n):         
        if n is not None:
            self.postorder1(n.left)
            self.postorder1(n.right)
            print(n.item, end=' ')

 
    def nodeCount(self):                 #노드 수 계산
        return self.nodeCount1(self.root)
        

    def nodeCount1(self,n):
        if n is None:
            return 0
        else:
            l=self.nodeCount1(n.left)
            r=self.nodeCount1(n.right)
            return 1+l+r

    
    def leafCount(self):                 #리프노드 수 계산
        return self.leafCount1(self.root)


    def leafCount1(self,n):  
        if n is None:
            return 0
        elif n.left is None and n.right is None:
            return 1
        else:
            return self.leafCount1(n.left)+self.leafCount1(n.right)


    def height(self):
        return self.height1(self.root)

    def height1(self,n):         #트리의 높이
        if n is None:
            return 0

        hLeft=self.height1(n.left)
        hRight=self.height1(n.right)
        if(hLeft>hRight):
            return hLeft+1
        else:
            return hRight+1



    def print(self):   #RVL 순회
        print()
        self.print1(self.root,0)
        
        
    def print1(self,n,count):
        
        if n is not None:
            count+=1
            self.print1(n.right,count)
            
           
            if n.right is not None and n.left is None:
                print(end='\t'*count )
                print(n.item,",R")
                #count+=1

            elif n.right is None and n.left is not None:
                print(end='\t'*count )
                print(n.item,",L")
                #count+=1
                
            elif n.right is not None and n.left is not None:
                print(end='\t'*count )
                print(n.item,",L,R")
                #count+=1
                
            else:
                #print(end='\t')
                print(end='\t'*count)
                print(n.item)
                #count+=1
                

            #print(end='\t')
        
            
            self.print1(n.left,count)
            count+=1
                          
            
            #     #print(end='\n')
            # if n.right is not None:
                
            #     self.print1(n.right)
            #     print(end='\t')

            # if n.item==self.root:
            #     print(n.item)
            # else:
                
            #     print(n.item)
            #     print(end='\t')
                
            
            # if n.left is not None:
            #     print(end='\t')
            #     self.print1(n.left)
        
            
            
        
        
        #print(end='\t')




            
            
            

```
