```python
class Bag:
    def __init__(self):
        self.items = []

# 구현 해야할 멤버 함수 
    def contains(self, e):
        if e in self.items:
            return "True"
        else:
            return "False"
        
    def insert(self, e):
        self.items.append(e)
    
    def remove(self, e):
        if e in self.items:
            self.items.remove(e)
    
    def count(self):
        return len(self.items)
    
    def __str__(self):
        return self.items

myBag = Bag()
myBag.insert("휴대폰")
myBag.insert("지갑")
myBag.insert("손수건")
myBag.insert('빗')
myBag.insert('연필')
print("가방속 물건: ", myBag.items)
myBag.__str__()
myBag.insert('빗')
myBag.remove('손수건')
myBag.insert('자료 구조 책')
print("가방속 물건: ", myBag.items)
print("가방속 물건 갯수: %d" %myBag.count())
print("가방속 지갑 유무: %s" %myBag.contains('지갑'))
print("가방속 손수건 유무: %s" %myBag.contains('손수건'))
```

# 문제 2 

2장에서 리스트에 있는 데이터의 최솟값, 최댓값을 찾는 findMinMax() 함수를 공부하였다. 이 함수를 이용하여 파일 “input.txt”에 있는 연, 월, 일로 주어진 날짜를 읽어 출력하고 가장 오래된 날짜와 가장 늦은 날짜를 출력하도록 한다.


```python
#제출용
class Date:
    def __init__(self, year,month, day):
        self.year=year
        self.month=month
        self.day=day
    
    #구현해야 할 멤버 함수
    def __gt__(self, rhs):
        return self.year>rhs.year
    
        
    def __lt__(self, rhs):
        return self.year<rhs.year

    
        
    def __str__(self):
        return "{0}/{1}/{2}".format(self.year, self.month, self.day)   #매직메서드 정의
    
            
#수업시간 배운 함수 사용
#리스트에 있는 날짜에서 가장 오래된 날짜와 가장 늦은 날짜를 출력
def findMinMax(lst):
    
    min=lst[0]
    max=lst[0]
    for i in range(1, len(lst)):
        if max<lst[i] : max=lst[i]
        if min>lst[i] : min=lst[i]
    return (min, max)                


#주어진 파일을 읽어 날짜를 리스트에 저장    
def main():
    inFile=open("input.txt", "r")
    lst=[]
    while True:
        line=inFile.readline()
        if line=="":
            break
            
        date=[int(i) for i in line.split()]
        print(date)
        lst.append(Date(date[0], date[1], date[2]))   #yyyy/mm/dd 형테로 lst에 저장됨
        
    for i in range(len(lst)):                      
        print(lst[i])                                #lst를 fimdMinMax에 넣음
        (min, max)=findMinMax(lst)
    print()
    print('earlist date: ', min)
    print('latest date: ', max)
    inFile.close()

main()        
```

# 문제3 
기준 날짜에서 날짜 수만큼 진행한 후의 날짜를 출력한다. 하루를 증가시키는 멤버 함수 increment()를 구현하여 사용한다. 필요하면 보조 멤버 함수를 추가하여 사용한다.

```python
#제출용
class Date:
    def __init__(self, year,month, day, afterday):
        self.year=year
        self.month=month
        self.day=day
        self.afterday=afterday
        

        
    def __str__(self):
        return "{0}/{1}/{2}".format(self.year, self.month, self.day, self.afterday)   #매직메서드 정의     
    
    def lastDayOfTheMonth(self):
        if self.month in [1,3,5,7,8,10,12]:
            return 31
        if self.month in [4,6,9,11]:
            return 30
        if self.month==2 and self.year%4==0:
            return 29
        if self.month==2 and self.year%4!=0:
            return 28
            
            
    def increment(self):
        if self.day == self.lastDayOfTheMonth():
            self.day = 1
            if self.month == 12:
                self.month = 1
                self.year += 1
            else:
                self.month += 1
        else:
            self.day += 1
        return self

def main():
    inFile=open("input2.txt", "r")
    lst=[]
    lst2=[]
    datebox=[]
    while True:
        line=inFile.readline()
        if line=="":
            break
            
        date=[int(i) for i in line.split()]
        box=Date(date[0], date[1], date[2], date[3])
        box2=Date(date[0], date[1], date[2], date[3])
        datebox.append(date[3])
        lst.append(box)
        
        for i in range(date[3]):
            box2=box2.increment()
            
        lst2.append(box2)   #yyyy/mm/dd 형테로 lst에 저장됨
        
    for i in range(len(lst)):
        print("{} {}일 후 ==> {}".format(lst[i],datebox[i],lst2[i])) 
        
        
    inFile.close()

main()    
    
```

# 문제 4
첫 번째 주어진 날짜에서 두 번째 날짜까지의 경과한 날짜 수를 셈하여 출력한다. 문제 3에서 사용한 하루를 증가시키는 멤버 함수 increment()를 이용하여 구한다.


```python
class Date:
    def __init__(self, year,month, day):
        self.year=year
        self.month=month
        self.day=day
    #구현해야 할 멤버 함수
    def __gt__(self, rhs):
        if self.year>rhs.year:
            return self.year>rhs.year
        if self.month>rhs.month:
            return self.year>rhs.year
        if self.day>rhs.day:
            return self.day>rhs.day       
    
        
    def __lt__(self, rhs):
        if self.year<rhs.year:
            return self.year<rhs.year
        if self.month<rhs.month:
            return self.year<rhs.year
        if self.day<rhs.day:
            return self.day<rhs.day
        
    def __eq__(self, rhs):
        if self.year==rhs.year:
            return self.year==rhs.year
        if self.month==rhs.month:
            return self.year==rhs.year
        if self.day==rhs.day:
            return self.day==rhs.day
        
        
    def __str__(self):
        return "{0}/{1}/{2}".format(self.year, self.month, self.day)   #매직메서드 정의     
    
    def lastDayOfTheMonth(self):
        if self.month in [1,3,5,7,8,10,12]:
            return 31
        if self.month in [4,6,9,11]:
            return 30
        if self.month==2 and self.year%4==0:
            return 29
        if self.month==2 and self.year%4!=0:
            return 28
            
            
    def increment(self):
        if self.day == self.lastDayOfTheMonth():
            self.day = 1
            if self.month == 12:
                self.month = 1
                self.year += 1
            else:
                self.month += 1
        else:
            self.day += 1
        
```

```python
#구상한 알고리즘

#작은 수가 큰 수가 될때까지 increment()를 호출한다.
#호출한 횟수만큼 지난 일수를 프린트한다.

#크기 비교
#년도가 큰가? 달이 큰가? 일이 큰가?
###왼쪽 오른쪽 구분하지 않고 그냥 같은지 다른지만 구분하면 간편해질듯

#while 같아질때까지

#왼쪽이 크면->오른쪽과 같아질 때까지 increment를 호출한다
#오른쪽이 크면->왼쪽과 같아질 때까지 increment를 호출한다
#같으면->0프린트


def main():
    inFile=open("input3.txt", "r")
    lst=[]
    lst2=[]
    
    while True:
        line=inFile.readline()
        if line=="":
            break
            
        date=[int(i) for i in line.split()]
        box=Date(date[0], date[1], date[2])
        box_copy=Date(date[0], date[1], date[2])
        
        box2=Date(date[3], date[4], date[5])
        box2_copy=Date(date[3], date[4], date[5])
        
        

    while True:
        count=0
        if (date[0]==date[3] and date[1]==date[4] and date[2]==date[5]):
            print("0")
            break

        if date[0]>date[3]:     #년비교
                count+=1               
                box2_copy.increment()

        else:
            if date[1]>date[4]:    #월비교
                count+=1               
                box2_copy.increment()

            elif date[2]>date[5]:   #일비교
                count+=1            
                box2_copy.increment()

            else:
                count+=1
                box_copy.increment()

            print("최종 {} {} {}".format(box,box2, count))
            print(" ")
        
#lst를 fimdMinMax에 넣음
    inFile.close()

main()  
```
