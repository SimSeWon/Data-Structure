```python
class DNode:
    def __init__(self, item, prev=None, next=None):
        self.item=item
        self.prev=prev
        self.next=next
```

```python
class CircularDoublyLinkedList:
    def __init__(self):
        self.head=None
        

    def isEmpty(self):return self.head==None
    def clear(self):self.head=self.rear=None
    
    def addFront(self, item):
        newNode=DNode(item)
        if (self.isEmpty()):
            self.head=newNode
            self.head.prev=newNode
            self.head.next=newNode

        else:
            newNode.prev=self.head.prev
            newNode.next=self.head
            self.head.prev.next=newNode
            self.head.prev=newNode
            self.head=newNode
            

    def addRear(self, item):
        newNode=DNode(item)
        if(self.isEmpty()):
            self.head=newNode
            self.head.prev=newNode
            self.head.next=newNode
        else:
            newNode.prev=self.head.prev
            newNode.next=self.head
            self.head.prev=newNode
            self.head.prev.next=newNode


    def deleteFront(self):
        if not self.isEmpty():
            item=self.head.item
            self.head=self.head.next
            if self.head==None:
                self.rear=None
            else:
                self.head.prev=None
            return item

    def deleteRear(self):
        if not self.isEmpty():
            item=self.rear.item
            self.rear=self.rear.prev
            if self.rear==None:
                self.head=None
            else:
                self.rear.next=None
            return item

    def peekFront(self):
        if not self.isEmpty():
            return self.head.item
    def peekRear(self):
        if not self.isEmpty():
            return self.rear.item
    def size(self):
        if self.head is None:
            return 0
        count=1
        tmp=self.head
        while tmp is not self.head.prev:
            count+=1
            tmp=tmp.next
        return count
        
    def print(self):
        newNode=self.head
        while not newNode==self.head.prev:
            print(newNode.item, end=' ')
            newNode=newNode.next
            
        
        print()

    def revPrint(self):       #역순출력
        newNode=self.rear
        while not newNode==None:
            print(newNode.item, end=' ')
        print()
```

```python
def main():
    cl=CircularDoublyLinkedList()
    print("Enter a command: af(addFront, df(deleteFront), pf(peekFront), s(size)")
    print("ar(addRear), dr(deleteRear), pr(peekRear), rp(reversePrint), q(uit)")

    while True:
        print("> ", end='')
        line=input().split()
        command=line[0]
        if command=='af':
            item=line[1]
            cl.addFront(item)
        elif command=='df':
            print(cl.deleteFront())
        elif command=='pf':
            print(cl.peekFront())
        elif command=='ar':
            item=line[1]
            cl.addRear(item)
        elif command=='dr':
            print(cl.deleteRear())
        elif command=='pr':
            print(cl.peekRear())    
        elif command=='p':
            cl.print()    
        elif command=='rp':
            cl.revPrint()
        elif command=='s':
            print("size: ", cl.size())
        elif command=='q':
            break        

```

```python
main()
```
